# Word2Vec

```python
from nltk.tokenize import word_tokenize
from sklearn.ensemble import RandomForestClassifier
from gensim.models import Word2Vec
import gensim
```

```python
raw_df
```

```python
df['target'] = df['sentiment'].apply(lambda x:1 if x=='positive' else 0)
```

```python
df['target']
```

```python
df['target'].shape
```

```python
df['review'].shape
```

```python
def preprocess_text(text):
  text = re.sub(r"[^a-zA-Z]", " ", text.lower())
  tokens = word_tokenize(text)
  return tokens
```

```python
df['tokens'] = df['review'].apply(preprocess_text)
```

```python
df['tokens']
```

```python
w2v_model = Word2Vec(sentences=df['tokens'],vector_size=100,window=5,min_count=2,workers=4)
```

```python
def get_vector(text_tokens, model):
  vectors = [model.wv[word] for word in text_tokens if word in model.wv]
  if vectors:
    return np.mean(vectors, axis=0)
  else:
    return np.zeros(model.vector_size)
```

```python
df['vector'] = df['tokens'].apply(lambda x: get_vector(x, w2v_model))
```

```python
df['vector']
```

# Train the RandomForestClassifier

```python
x = np.vstack(df['vector'].values)
y = df['target'].values
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
clf = RandomForestClassifier()
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
conf_matrix = confusion_matrix(y_test,y_pred)
print("Confution Matrix:\n", conf_matrix)
```

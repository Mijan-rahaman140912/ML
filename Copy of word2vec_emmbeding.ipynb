```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
import nltk
nltk.download('punkt_tab')
```

```python
story = []
for filename in os.listdir('game-of-thrones-books'):
  if filename == '.ipynb_checkpoints':
    pass
  # Check if item is a file before attempting to open it
  elif os.path.isfile(os.path.join('game-of-thrones-books', filename)):
    # Open the file in binary read mode ('rb') to read the raw bytes
    with open(os.path.join('game-of-thrones-books', filename), 'rb') as f:
      # Read the raw bytes
      corpus_bytes = f.read()
      # Try to decode using 'latin-1' or other encoding
      corpus = corpus_bytes.decode('latin-1')  # Replace 'latin-1' if it doesn't work
    rwa_sent = sent_tokenize(corpus)
    for sent in rwa_sent:
      story.append(simple_preprocess(sent))
```

```python
len(story)
```

```python
story[0]
```

```python
model = gensim.models.Word2Vec(
    window=10,
    min_count=2,

)
```

```python
model.build_vocab(story)
```

```python
model.train(story, total_examples=model.corpus_count, epochs=model.epochs)
```
